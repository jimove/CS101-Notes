# SymPy
-------

`SymPy`可以直观地展现原始的数学表达式

```python
import sympy #引入sympy库

x = sympy.S('x') #将变量设置为符号

x ** 2
```

-----

但是必须使用`SymPy`内置的数学函数

```python
sympy.exp(-x**2)
```

```python
sympy.sin(sympy.pi/4)
```

------

用`print()`输出回归原来的一行表达式

```python
print(sympy.sin(sympy.pi/4))
```

-----------

### 用SymPy解方程

```python
import sympy
a = sympy.S( 'a' )
sympy.solve(sympy.tan(a) - 2)
```

```python
a = sympy.S( 'a' )
b = sympy.S( 'b' )
c = sympy.S( 'c' )
x = sympy.S( 'x' )

sympy.solve( a*x**2+b*x+c, x )
```

-----

还可以变换主元

```python
sympy.solve( a*x**2+b*x+c,b )
```

--------------

### 因式分解、展开和组合

```python
sympy.factor( x**3 + x**2 - 4*x - 4 ) #因式分解
```

```python
sympy.expand( (x+1)*(x-1) ) #展开
```

```python
sympy.together( b/c+x/a ) #组合
```

-------

### 用SymPy绘图

```python
sympy.plotting.plot( x**2 ) #一维
```

```python
sympy.plotting.plot(x**2, (x, -2, 2)) #设置参数和范围
```

```python
x,y = sympy.S( 'x,y' )
sympy.plotting.plot3d( sympy.cos( x ) * sympy.sin( y ) )  #三维图像
```

```python
t = sympy.S( 't' )
sympy.plotting.plot3d_parametric_line( sympy.cos( t ),sympy.sin( t ),t )  #三维曲线
```

```python
sympy.plotting.plot3d_parametric_line( sympy.cos(x),sympy.sin(x),x,( x,-5,5 ) ) #设置参数和范围
```

```python
u,v = sympy.S( 'u,v' )
sympy.plotting.plot3d_parametric_surface( ( 3 + sympy.cos( u ) ) * sympy.cos( v ),
                                          ( 3 + sympy.cos( u ) ) * sympy.sin( v ),
                                          sympy.sin( u ) + 0.5 * v )
#三维曲面
```
